```python
import pandas as pd
df = pd.read_csv("cleaned_retail_dataset.csv")
```

```python
top_5_cities = df.groupby('city')['net_amount'].sum().sort_values(ascending=False).head(5)
print(f"1. Top 5 Cities by Revenue:\n{top_5_cities}")
```

```python
top_10_products = df.groupby('product_name')['net_amount'].sum().sort_values(ascending=False).head(10)
print(f"2. Top 10 Products by Sales:\n{top_10_products}")
```

```python
most_profitable_category = df.groupby('category')['net_amount'].sum().sort_values(ascending=False).head(1)
print(f"3. Most Profitable Category:\n{most_profitable_category}")
```

```python
avg_delivery_time = df.groupby('city')['delivery_days'].mean().sort_values()
print(f"4. Average Delivery Time by City (Days):\n{avg_delivery_time}")
```

```python
payment_counts = df['payment_method'].value_counts()
payment_percentage = df['payment_method'].value_counts(normalize=True) * 100

print("5. Payment Method Distribution:")
print(pd.DataFrame({'Count': payment_counts, 'Percentage (%)': payment_percentage.round(2)}))
```

```python
print("6. Identified Business Problems:\n")

# Problem 1: Revenue Concentration
# Calculate share of top 3 cities
top_3_share = top_5_cities.head(3).sum() / df['net_amount'].sum() * 100
print(f"   [Problem 1] Revenue Concentration Risk:")
print(f"      - The top 3 cities account for {top_3_share:.2f}% of total revenue.")
print(f"      - Insight: The business is highly dependent on a few locations.")

# Problem 2: Delivery Consistency
# Check variance in delivery times
delivery_std = df.groupby('city')['delivery_days'].mean().std()
if delivery_std < 1:
    print(f"\n   [Problem 2] Lack of Express Delivery Options:")
    print(f"      - Average delivery time is ~{df['delivery_days'].mean():.1f} days across all cities with low variance.")
    print(f"      - Insight: No faster delivery options for premium customers or major cities.")

# Problem 3: Cash Dependency
cash_share = payment_percentage.get('Cash', 0)
print(f"\n   [Problem 3] High Operational Risk in Payments:")
print(f"      - Cash payments make up {cash_share:.2f}% of transactions.")
print(f"      - Insight: High reliance on Cash on Delivery (COD) increases risk of returns and theft.")
```
